```python
import numpy as np                                  # numeric operations
import matplotlib.pyplot as plt                     # plotting library
np.random.seed(42)                                  # reproducible randomness

m = 100                                             # number of training examples
X = 2 * np.random.rand(m, 1)                        # feature values (m x 1)
true_theta0 = 4.0                                   # true intercept
true_theta1 = 3.0                                   # true slope
noise = np.random.randn(m, 1) * 1.0                 # gaussian noise term
y = true_theta0 + true_theta1 * X + noise          # target values (m x 1)
# --- add bias (column of ones) to X to get design matrix ---
X_b = np.c_[np.ones((m, 1)), X]                     # design matrix (m x 2)
# --- helper function: compute cost J(theta) for linear regression ---
def compute_cost(theta, X, y):                      # theta should be shaped (2,1)
    m = len(y)                                      # number of examples
    predictions = X.dot(theta)                      # hypothesis h_theta(x) = X * theta
    errors = predictions - y                        # prediction errors (m x 1)
    cost = (1 / (2 * m)) * np.sum(errors ** 2)      # mean squared error cost (scalar)
    return float(cost)                              # return scalar cost value
# --- gradient descent implementation (batch) ---
def gradient_descent(X, y, initial_theta, learning_rate, n_iters):  # signature
    m = len(y)                                      # number of training examples
    theta = initial_theta.copy()                    # copy initial parameters to avoid side-effects
    cost_history = []                               # list to store cost at each iteration
    theta_history = []                              # list to store flattened theta pairs (for plotting)
    for i in range(n_iters):                        # loop for n_iters gradient steps
        predictions = X.dot(theta)                  # compute current predictions (m x 1)
        errors = predictions - y                    # compute errors (m x 1)
        gradients = (1 / m) * X.T.dot(errors)       # gradient vector (2 x 1)
        theta = theta - learning_rate * gradients   # parameter update: theta := theta - alpha * grad
        cost = compute_cost(theta, X, y)           # compute cost after update
        cost_history.append(cost)                   # append cost scalar to history list
        theta_history.append([float(theta[0]), float(theta[1])])  # append flattened theta pair
    theta_history = np.array(theta_history)         # convert list to numpy array shape (n_iters, 2)
    return theta, np.array(cost_history), theta_history  # return final theta, cost history, theta history
# --- run gradient descent with chosen hyperparameters ---
initial_theta = np.zeros((2, 1))                    # initialize theta as zeros (2 x 1)
learning_rate = 0.1                                  # step size alpha
n_iters = 1000                                      # number of iterations to run
final_theta, cost_hist, theta_hist = gradient_descent(X_b, y, initial_theta, learning_rate, n_iters)  # run
# --- print final results ---
print("Final parameters (theta0, theta1):", final_theta.ravel())  # learned intercept and slope
print("True parameters (theta0, theta1):", np.array([true_theta0, true_theta1]))  # ground truth values
print("Final cost:", cost_hist[-1])                    # last cost value after training
# --- plot 1: data points and fitted line ---
plt.figure()                                         # new figure for scatter + line
plt.scatter(X, y)                                    # scatter data points
x_line = np.array([[0.0], [2.0]])                    # two x-values to draw fitted line across
x_line_b = np.c_[np.ones((2, 1)), x_line]            # add bias term to x-line (2 x 2)
y_line = x_line_b.dot(final_theta)                   # predicted y values for the line (2 x 1)
plt.plot(x_line, y_line)                             # plot fitted regression line
plt.xlabel("x (feature)")                            # label x-axis
plt.ylabel("y (target)")                             # label y-axis
plt.title("Data and Learned Linear Fit")             # title for the figure
plt.show()                                           # display the plot                                       # display the plot
# --- plot 3: contour of cost surface and theta path ---
theta0_vals = np.linspace(final_theta[0] - 10, final_theta[0] + 10, 100)  # range for intercept
theta1_vals = np.linspace(final_theta[1] - 10, final_theta[1] + 10, 100)  # range for slope
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))  # container for cost values
for i, t0 in enumerate(theta0_vals):                  # loop over theta0 values
    for j, t1 in enumerate(theta1_vals):              # loop over theta1 values
        t = np.array([t0, t1]).reshape(2, 1)          # ensure theta is shape (2,1) for compute_cost
        J_vals[i, j] = compute_cost(t, X_b, y)       # compute cost and store it
T0, T1 = np.meshgrid(theta0_vals, theta1_vals)        # grid for contour plotting
plt.figure()                                          # new figure for contour
plt.contour(T0, T1, J_vals.T, levels=30)              # draw contour lines of the cost surface
plt.plot(theta_hist[:, 0], theta_hist[:, 1], marker='o', markersize=3, linewidth=1)  # overlay parameter path
plt.xlabel("theta0 (intercept)")                      # x-axis label
plt.ylabel("theta1 (slope)")                          # y-axis label
plt.title("Cost Contours and Gradient Descent Path")  # figure title
plt.show()                                            # display the plot
# --- print a short numeric snapshot of the first 5 theta pairs and first 5 costs ---
print()
print("First 5 theta pairs (theta0, theta1):")
print(theta_hist[:5])
print("First 5 costs:")
print(cost_hist[:5])
```
