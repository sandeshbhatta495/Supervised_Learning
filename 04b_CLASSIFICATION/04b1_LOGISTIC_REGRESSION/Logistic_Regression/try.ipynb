```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(
    n_samples=100,
    n_features=2,
    centers=6,
    cluster_std=1,
    random_state=5
)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Blob Data Visualization")
plt.show()
```
