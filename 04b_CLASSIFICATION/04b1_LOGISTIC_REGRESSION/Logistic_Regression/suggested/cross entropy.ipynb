```python
import numpy as np

# Ground truth
y = np.array([1, 0, 1, 1])

# Predicted probabilities
y_hat = np.array([0.9, 0.1, 0.4, 0.8])

# Epsilon for numerical stability
epsilon = 1e-9
y_hat = np.clip(y_hat, epsilon, 1 - epsilon)

# Cross entropy loss
loss = -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

print("Cross Entropy Loss:", loss)
```

```python
# Loss ≈ 0.69  → random guessing  
# Loss ≈ 0.50  → learning
# Loss ≈ 0.33  → good model
# Loss < 0.20  → very strong model
# Loss → 0     → perfect (rare)
```
