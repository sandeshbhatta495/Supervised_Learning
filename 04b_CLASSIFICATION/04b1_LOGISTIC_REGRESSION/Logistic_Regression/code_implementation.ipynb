```python
# Libraies 
from sklearn.datasets import make_classification # for generating a random classification problem
import numpy as np
```

```python
X, y = make_classification(
    n_samples=100,            # total number of samples
    n_features=2,             # number of features
    n_informative=1,          # number of informative features
    n_redundant=0,            # no redundant features
    n_classes=2,              # binary classification
    n_clusters_per_class=1,   # one cluster for each class
    random_state=41,          # reproducibility
    hypercube=False,          # simple cluster shapes
    class_sep=10              # classes far apart
)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
