```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
X = data.data               
y = data.target   
feature_names = data.feature_names
```

# 0 = malignant, 1 = benign

```python
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y
```

```python
df.shape
```

```python
df.info
```

```python
df.describe()
```

```python
# How many samples of each class?
print("\nClass distribution:")
print(df['target'].value_counts())   # 0 malignant, 1 benign
```

```python
# Train Test spilt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_features = df.drop('target', axis=1).values
y_target = df['target'].values
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y_target, test_size=0.2, random_state=42, stratify=y_target # 20% of data are going to be gone to the test the model 
)
```

```python
# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
X_test_scaled.shape
```

```python
df.head()
```

# using KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

# checking the accuracy Based on the value of K ---------

```python
scores = []
for i in range(1,16):   
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred)) 
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,16),scores)
```
