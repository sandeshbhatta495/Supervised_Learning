```python
# libraries 
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
X = data.data                # shape (n_samples, 30)
y = data.target              # 0 = malignant, 1 = benign
feature_names = data.feature_names
```

```python
# Create DataFrame for easy inspection
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y
```

```python
print("Shape:", df.shape)
print("\nFeature names (30):")
print(feature_names)
```

```python
# Descriptive statistics (mean, std, min, max)
df.describe()
```

```python
display(df.describe()) 
```

```python
# How many samples of each class?
print("\nClass distribution:")
print(df['target'].value_counts())   # 0 malignant, 1 benign
```

```python
# compare distribution of one important feature by class
import matplotlib.pyplot as plt
# plt.figure(figsize=(6,4))
plt.hist(df[df['target']==0]['mean radius'], bins=20, alpha=0.6, label='malignant')
plt.hist(df[df['target']==1]['mean radius'], bins=20, alpha=0.6, label='benign')
plt.xlabel('mean radius')
plt.legend()
plt.show()
```

```python
# Train Test spilt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_features = df.drop('target', axis=1).values
y_target = df['target'].values

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y_target, test_size=0.2, random_state=42, stratify=y_target # 20% of data are going to be gone to the test the model 
)

# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Train shape:", X_train_scaled.shape)
print("Test shape:", X_test_scaled.shape)
```

# Train Logistic Regression Model (Main Part)

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, solver='lbfgs')
model.fit(X_train_scaled, y_train)

print("Model trained successfully")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_pred = model.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)

print("Accuracy:", acc)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
In  Mathmatical and therotical way
1. Accuracy
Accuracy measures total correct predictions out of total samples:
Accuracy= (TP+TN)/TP+TN+FP+FN
Where:

TP = True Positives

TN = True Negatives

FP = False Positives

FN = False Negatives
```

```python
# predict_proba gives the probability of each class for every test sample
# It returns two columns: 
# Column 0 → Probability of Malignant (class 0)
# Column 1 → Probability of Benign (class 1)
probabilities = model.predict_proba(X_test_scaled)

# Print the first 10 probability predictions to understand the format
print("First 10 Probability Predictions:\n", probabilities[:10])

# Extract only the probability for class 1 (Benign)
# This is the second column because index 1 = benign probability
benign_prob = probabilities[:, 1]

# Extract only the probability for class 0 (Malignant)
# This is the first column because index 0 = malignant probability
malignant_prob = probabilities[:, 0]

# Print the first 10 probabilities for benign
print("\nBenign Probabilities (first 10):\n", benign_prob[:10])

# Print the first 10 probabilities for malignant
print("\nMalignant Probabilities (first 10):\n", malignant_prob[:10])
```
