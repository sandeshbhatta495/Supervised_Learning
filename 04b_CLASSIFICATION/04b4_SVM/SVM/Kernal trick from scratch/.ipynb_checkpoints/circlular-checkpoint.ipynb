```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
np.random.seed(0)

# Class +1 (inner circle)
theta = np.linspace(0, 2*np.pi, 50)
r1 = 1
X1 = np.c_[r1*np.cos(theta), r1*np.sin(theta)]

# Class -1 (outer circle)
r2 = 3
X2 = np.c_[r2*np.cos(theta), r2*np.sin(theta)]

X = np.vstack((X1, X2))
y = np.hstack((np.ones(50), -np.ones(50)))
```

# Visualize Raw Data

```python
plt.scatter(X[y==1][:,0], X[y==1][:,1], label="+1")
plt.scatter(X[y==-1][:,0], X[y==-1][:,1], label="-1")
plt.legend()
plt.title("Non-linearly Separable Data")
plt.show()
```

```python
def rbf_kernel(x, z, gamma=0.5):
    return np.exp(-gamma * np.linalg.norm(x - z)**2)
```

```python
class KernelSVM:
    def __init__(self, lr=0.01, C=1, gamma=0.5, epochs=100):
        self.lr = lr
        self.C = C
        self.gamma = gamma
        self.epochs = epochs

    def fit(self, X, y):
        self.X = X
        self.y = y
        self.alpha = np.zeros(len(X))
        self.b = 0

        for _ in range(self.epochs):
            for i in range(len(X)):
                decision = sum(
                    self.alpha[j] * y[j] * rbf_kernel(X[j], X[i], self.gamma)
                    for j in range(len(X))
                ) + self.b

                if y[i] * decision < 1:
                    self.alpha[i] += self.lr
                    self.b += self.lr * y[i]

    def predict(self, X_test):
        preds = []
        for x in X_test:
            val = sum(
                self.alpha[i] * self.y[i] * rbf_kernel(self.X[i], x, self.gamma)
                for i in range(len(self.X))
            ) + self.b
            preds.append(np.sign(val))
        return np.array(preds)
```
