```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_circles
```

```python
# Generate non-linear data
X, y = make_circles(n_samples=200, factor=0.5, noise=0.05)
```

```python

# Fit SVM with RBF kernel
clf = SVC(kernel='rbf', gamma=50, C=1)
clf.fit(X, y)
```

```python
# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 500),
                     np.linspace(-1.5, 1.5, 500))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, levels=[-1,0,1], alpha=0.2, colors=['blue','red'])
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', edgecolors='k')
plt.show()
```

# In Support Vector Machines (SVM), particularly with nonlinear kernels like RBF, gamma controls the influence radius of a single training example; a high gamma means a small radius, making the decision boundary complex and prone to overfitting, while a low gamma means a large radius, creating a smoother boundary but potentially underfitting. It essentially defines how far the "shadow" or influence of a data point extends, determining the model's sensitivity to individual points. 
How gamma works:
High gamma: Each point's influence is localized to a small area. The model becomes very sensitive to individual points, potentially fitting the training data too closely (overfitting) and creating intricate, wiggly boundaries.
Low gamma: A single training point influences a large area. The decision boundary becomes smoother and more generalized, grouping distant points together (underfitting if too low). 
Analogy:
Think of gamma as setting the "reach" of each data point.
High gamma (small reach): Like a spotlight on one person; only very close people are affected.
Low gamma (large reach): Like a floodlight; many people, even those further away, are illuminated and considered part of the group.
