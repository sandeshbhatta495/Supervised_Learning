# Model Traning 

```python
X = df[["cgpa", "iq"]]
y = df["placement"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

# train test split 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

# MODEL BUILDING 

```python
rf = RandomForestClassifier(
    n_estimators=100,     # number of trees
    max_depth=5,          # depth of each tree
    min_samples_split=2,  # minimum samples to split
    max_features="sqrt",  # random feature selection
    bootstrap=True,       # enable bagging
    random_state=42
)
```

# Training  the model 

```python
rf.fit(X_train, y_train)
```

# Make Prediction 

```python
y_pred = rf.predict(X_test)
```

# Accuracy Score 

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

# Important feature Determining 

```python
importances = rf.feature_importances_

for feature, importance in zip(X.columns, importances):
    print(feature, ":", importance)
```

```python
Meaning
Higher value  feature contributes more to decision
Helps explain model behavior
```

# OOB Error Estimation

```python
rf_oob = RandomForestClassifier(
    n_estimators=100,
    oob_score=True,
    random_state=42
)

rf_oob.fit(X_train, y_train)

print("OOB Score:", rf_oob.oob_score_)
```

# Internal validation without test set

```python
# plotting the decision boundary 
from matplotlib.colors import ListedColormap
```

```python
def plot_decision_boundary(rf, X, y):
    X = X.values
    y = y.values

    x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
    y_min, y_max = X[:,1].min() - 5, X[:,1].max() + 5

    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, 0.02),
        np.arange(y_min, y_max, 1)
    )

    Z = rf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
    cmap_bold = ['red', 'green']
    
    plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
    plt.scatter(X[:,0], X[:,1], c=y, cmap=ListedColormap(cmap_bold), edgecolor='k')
    plt.xlabel("CGPA")
    plt.ylabel("IQ")
    plt.title("Random Forest Decision Boundary")
    plt.show()
```

```python
plot_decision_boundary(rf, X, y)
```

# Hyper parameter Tunning 

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 4, 6, 8],
    'min_samples_split': [2, 5, 10],
    'max_features': ['sqrt', 'log2']
}
```

```python
rf_base = RandomForestClassifier(random_state=42)

grid = GridSearchCV(
    rf_base,
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid.fit(X_train, y_train)
```
